```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

# Linearity Engineering 

```python
n_points = 100
radii = np.random.uniform(0, 6, n_points)
angles = np.random.uniform(0, 3 * np.pi, n_points)

# Classes: radius < 2.5 --> Class 0, else Class 1
labels = (radii > 2.5).astype(int)

# Convert to cartesian
x = radii * np.cos(angles)
y = radii * np.sin(angles)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=labels, cmap='cool', edgecolor='k', alpha=0.7)
plt.title("Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()

# Polar Coordinates
plt.figure(figsize=(8, 6))
plt.scatter(angles, radii, c=labels, cmap='cool', edgecolor='k', alpha=0.7)
plt.plot([2.45]*12)
plt.title("Data in Polar Coordinates")
plt.xlabel("Angle (θ)")
plt.ylabel("Radius (r)")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1, 100, n_points)
noise = np.random.normal(0, 0.2, size=x.shape)

# Generate y values
y = np.log(x) + noise

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='aqua', edgecolor='k', alpha=0.7)
plt.scatter(np.log(x), y, c='purple', edgecolor='k', alpha=0.7)
plt.title("Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

# Plot log(x)
plt.figure(figsize=(8, 6))
plt.scatter(np.log(x), y, c='purple', edgecolor='k', alpha=0.7)
plt.title("Transformed Data")
plt.xlabel("log(x)")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

# Activation functions Plot

```python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, x*alpha) 

def leaky_relu_derivative(x, alpha=0.1):
    return np.where(x > 0, 1, alpha)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

# Set up x values
x = np.linspace(-5, 5, 500)

# Set up plots
functions = [
    ("ReLU", relu, relu_derivative),
    ("Sigmoid", sigmoid, sigmoid_derivative),
    ("tanh", tanh, tanh_derivative),
    ("Leaky ReLU", leaky_relu, leaky_relu_derivative),
]

# Create a folder to save plots
output_folder = "plots"
# os.makedirs(output_folder, exist_ok=True)

for name, func, deriv in functions:
    plt.figure(figsize=(6, 4))
    plt.plot(x, func(x), label=f'{name}', color='blue')
    plt.plot(x, deriv(x), linestyle='--', label=f'{name} Derivative', color='red')
    plt.title(name)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # filename = os.path.join(output_folder, f"{name.lower()}.png")
    # plt.savefig(filename)
    plt.close()

# print(f"Plots saved in the '{output_folder}' folder!")
```
